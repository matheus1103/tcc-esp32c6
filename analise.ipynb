### Gerar strings aleatorias

```python
import random
import string

# Função para gerar strings aleatórias
def gerar_string_aleatoria(tamanho):
    caracteres = string.ascii_letters + string.digits
    return ''.join(random.choice(caracteres) for _ in range(tamanho))

# Gerar 100 strings aleatórias com tamanhos variáveis (até 32)
strings_aleatorias = [gerar_string_aleatoria(random.randint(64, 640)) for _ in range(100)]

# Salvar as strings em um array
array_de_strings = strings_aleatorias
array_de_strings
```

### análise RSA com LOG

```python
import pandas as pd

def ms_para_minuto_segundo(ms):
    minutos = ms // 60000
    segundos = (ms % 60000) // 1000
    return f"{minutos}m {segundos}s"

def formatar_corrente(corrente):
    return f"{corrente:.2f}mA"

# Criar um DataFrame com os logs
df_corrente = pd.read_csv('RSA_corrente_com_log.csv')

# Filtrar para valores Corrente(mA) > 26
df_filtrado = df_corrente[abs(df_corrente['Corrente(mA)']) > 25]
tempo_execucao = df_filtrado['Timestamp(ms)'].iloc[-1] - df_filtrado['Timestamp(ms)'].iloc[0]

df_filtrado['Timestamp(ms)'] = df_filtrado['Timestamp(ms)'].apply(ms_para_minuto_segundo)

# Exibir linhas do DataFrame filtrado
df_filtrado.tail()
```

```python
# tempo total de execução (subtrair pelo primeiro valor)
tempo_total = ms_para_minuto_segundo(tempo_execucao)
tempo_total
```

```python
# Valor medio da corrente
media_corrente = df_filtrado['Corrente(mA)'].mean()
media_corrente
```
